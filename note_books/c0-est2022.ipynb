```python
#data from https://www.census.gov/data/datasets/time-series/demo/popest/2020s-counties-total.html

# import pandas lib
import pandas as pd
import matplotlib.pyplot as plt
#get data
data_file = '/Users/chrisv/data_gov/data/co-est2022-alldata.csv'

#read into pandas
data = pd.read_csv(data_file)

#check data loaded, use the describe method so that we can see some genrated stats on the data
data.describe()
```

```python

#SELECT rows where COLOMN IS VALUE


data.loc[data['STNAME'] == 'Alabama']
```

```python
#filter columns a key word
filter_col = [col for col in data if col.startswith('S')]
data[filter_col]
```

```python
#Filter rows in a specific column that start with 'C'
#loc accesses a group of rows and columns
data.loc[data['CTYNAME'].str.startswith('Cof', na=False)]
```

```python
#data by slicing 
data.iloc[:, 1:4]
```

```python
#get counts of each state using value_counts()
data['STNAME'].value_counts()
```

```python
#basic methods, just an example.
city_count = data.groupby(by='STNAME').CTYNAME.count()
mean = city_count.mean()
median = city_count.median()
std = city_count.std()
print(f"Mean: {mean}, Median: {median}, Standard Deviation: {std}")
```

```python
#group data by state and est 2020 pop 
grouped_data = data.groupby('STNAME')['POPESTIMATE2020'].sum().reset_index()
grouped_data = grouped_data.sort_values(by='POPESTIMATE2020', ascending=False)
grouped_data
```

```python


plt.figure(figsize=(12, 8)) 

# bar plot 
plt.bar(grouped_data['STNAME'], grouped_data['POPESTIMATE2020'], color='skyblue')

#rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

#add labels and title
plt.xlabel('State')
plt.ylabel('Population Estimate 2020')
plt.title('Population Estimate 2020 per State')

#plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 12)) 
# horizontal bar plot 
plt.barh(grouped_data['STNAME'], grouped_data['POPESTIMATE2020'], color='skyblue')

#rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

#add labels and title
plt.xlabel('Population Estimate 2020')
plt.ylabel('State')
plt.title('Population Estimate 2020 per State')

#plot
plt.tight_layout()
plt.show()
```

```python
#sum all births per state
grouped_birth_data = data.groupby('STNAME')['BIRTHS2020'].sum().reset_index()
grouped_birth_data = grouped_birth_data.sort_values(by='BIRTHS2020', ascending=False)
grouped_birth_data
```

```python
#state birth 2020 using column BIRTHS2020

plt.figure(figsize=(10, 12)) 
# horizontal bar plot 
plt.barh(grouped_birth_data['STNAME'], grouped_birth_data['BIRTHS2020'], color='skyblue')

#rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

#add labels and title
plt.xlabel('State Births 2020')
plt.ylabel('State')
plt.title('Briths per State in 2020')

#plot
plt.tight_layout()
plt.show()
```
